```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
cleaned_data_count
```

```python
ax = recent_grads.hist("Sample_size")
plt.show()
```

```python
ax = recent_grads.plot.scatter("Sample_size", "Median")
ax.set_title("Median vs Sample_size")
ax.set_xlim(0, 500)
plt.show()
```

```python
recent_grads.plot.scatter("Sample_size", "Unemployment_rate")
plt.show()
```

```python
recent_grads.plot.scatter("Full_time", "Median")
plt.show()
```

```python
recent_grads.plot.scatter("ShareWomen", "Unemployment_rate")
plt.show()
```

```python
recent_grads.plot.scatter("Men", "Median")
plt.show()
```

```python
recent_grads.plot.scatter("Women", "Median")
plt.show()
```

```python
columns = ["Sample_size","Median","Employed","Full_time","ShareWomen","Unemployment_rate","Men","Women"]
for column in columns:
    recent_grads[column].plot(kind="hist", bins=25, rot=40)
    plt.show()
```
