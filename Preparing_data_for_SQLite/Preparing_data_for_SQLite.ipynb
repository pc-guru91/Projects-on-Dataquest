```python
academy_awards['Year'] = academy_awards['Year'].str[:4].astype('int64')
```

```python
later_than_2000 = academy_awards[academy_awards.Year > 2000]
```

```python
award_categories = ['Actor -- Leading Role', 'Actor -- Supporting Role', 'Actress -- Leading Role', 'Actress -- Supporting Role']
```

```python
nominations = later_than_2000[later_than_2000.Category.isin(award_categories)]
```

```python
nominations.loc[:,'Won?'] = nominations.loc[:,'Won?'].map({'NO': 0, 'YES': 1}).astype('int64')
```

```python
nominations.loc[:,'Won'] = nominations.loc[:,'Won?']
```

```python
final_nominations = nominations.drop(['Won?', 'Unnamed: 5',
       'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], axis=1)
```

```python
final_nominations.head()
```

```python
final_nominations[['Movie','Character']] = final_nominations['Additional Info'].str.extract('(^.+)\{\'(.+)\'\}', expand=False)
```

```python
final_nominations.drop('Additional Info', axis=1, inplace=True)
```

```python
query = "CREATE TABLE IF NOT EXISTS ceremonies(id integer PRIMARY KEY, Year integer, Host text)"
conn.execute(query)
```

```python
years_hosts = [(2010, "Steve Martin"),
               (2009, "Hugh Jackman"),
               (2008, "Jon Stewart"),
               (2007, "Ellen DeGeneres"),
               (2006, "Jon Stewart"),
               (2005, "Chris Rock"),
               (2004, "Billy Crystal"),
               (2003, "Steve Martin"),
               (2002, "Whoopi Goldberg"),
               (2001, "Steve Martin"),
               (2000, "Billy Crystal"),
            ]
insert_query = "INSERT INTO ceremonies (Year, Host) VALUES (?,?);"
conn.executemany(insert_query, years_hosts)
query = "SELECT * FROM ceremonies LIMIT 10"
pd.read_sql_query(query, conn)
```

```python
query = '''
SELECT nominations.category, nominations.nominee, nominations.movie, nominations.character, nominations.won, ceremonies.id
FROM nominations
INNER JOIN ceremonies ON
nominations.year == ceremonies.year
'''
joined_nominations = list(conn.execute(query))
```

```python
query = '''
CREATE TABLE IF NOT EXISTS nominations_two(id integer primary key, category text, nominee text,
movie text, character text, won integer, 
ceremony_id integer references ceremonies(id))
'''
conn.execute(query)
```

```python
query = "PRAGMA foreign_keys = ON"
conn.execute(query)
insert_query = '''
INSERT INTO nominations_two (category, nominee, movie, character, won, ceremony_id) VALUES (?, ?, ?, ?, ?, ?)
'''
conn.executemany(insert_query, joined_nominations)
```
